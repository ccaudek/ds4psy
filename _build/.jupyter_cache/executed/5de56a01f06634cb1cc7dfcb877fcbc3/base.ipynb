```python
theta = np.linspace(0, 1, 200)
alpha = 25
beta = 15
pdf = stats.beta.pdf(theta, alpha, beta)
plt.plot(theta, pdf, label=r"$\alpha$ = {}, $\beta$ = {}".format(alpha, beta))
plt.xlabel(r"$\theta$", fontsize=14)
plt.ylabel("Densit di probabilit", fontsize=14)
plt.legend(loc=1)
plt.show()
```

```python
# Parametri della distribuzione Beta
alpha = 25
beta = 15

# Calcoliamo la probabilit P(theta < 0.5) utilizzando la funzione cdf 
probability = stats.beta.cdf(0.5, alpha, beta)

# La probabilit P(theta < 0.5)  data da 1 - P(theta > 0.5)
probability_less_than_0_5 = 1 - probability

print(f"La probabilit P(theta < 0.5) per una Beta(25, 15) : {probability_less_than_0_5:.4f}")
```
